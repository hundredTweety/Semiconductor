```python
import matplotlib.pyplot as plt

# 박막 두께 열만 선택합니다.
thickness_columns = train.iloc[:, 0:5]

# 각 박막의 두께 분포를 히스토그램으로 시각화합니다.
thickness_columns.hist(bins=20, alpha=0.7,figsize=(16,16))
plt.xlabel('Thickness')
plt.ylabel('Frequency')
plt.title('Thickness Distribution of Thin Films')
plt.show()
```

```python
import seaborn as sns

# 박막 두께 데이터에 대한 산점도 행렬을 생성합니다.
sns.pairplot(thickness_columns)
plt.suptitle('Scatter Matrix of Thin Film Thicknesses', y=1.02)
plt.show()
```

```python
# 파장 역수와 반사율 스펙트럼 값을 선택합니다.
spectral_data = test.iloc[:, 5:]

# 첫 번째 샘플에 대한 반사율 스펙트럼을 시각화합니다. 추가 샘플은 루프를 사용하여 추가할 수 있습니다.
plt.plot(spectral_data.iloc[0], label='Sample 0')
plt.xlabel('Wavenumber')
plt.ylabel('Reflectance')
plt.title('Reflectance Spectrum')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
import numpy as np

# 반사율 스펙트럼 데이터를 PCA로 차원 축소합니다.
pca = PCA(n_components=3)  # 주성분을 2개로 설정합니다.
spectral_data_reduced = pca.fit_transform(spectral_data)

# 차원 축소 결과를 시각화합니다.
plt.scatter(spectral_data_reduced[:, 0], spectral_data_reduced[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Reflectance Spectrum Data')
plt.show()
```
